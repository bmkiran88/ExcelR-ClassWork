```python
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
load_breast_cancer()
```

```python
cancer_data=pd.DataFrame(load_breast_cancer().data,columns=load_breast_cancer().feature_names)
cancer_data['target']=pd.DataFrame(load_breast_cancer().target)
```

```python
cancer_data
```

```python
X=cancer_data.iloc[:,:-1]
y=cancer_data[['target']]
```

```python
X
```

```python
y
```

```python
pd.set_option('max_columns',None)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.title('Nature of the data',size=20)
sns.countplot(x=cancer_data.target)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=13)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=12,class_weight=None)
dt_model.fit(X_train,y_train)
```

```python
y_pred_train=dt_model.predict(X_train)
```

```python
y_pred_test=dt_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
confusion_matrix(y_test,y_pred_test)
```

## After giving weight to imbalance weightage

```python
dt_model=DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=12,class_weight={0:2,1:1})
dt_model.fit(X_train,y_train)
```

```python
y_pred_train=dt_model.predict(X_train)
y_pred_test=dt_model.predict(X_test)
```

```python
confusion_matrix(y_train,y_pred_train)
```

```python
confusion_matrix(y_test,y_pred_test)
```
