## 1. Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Importing Data

```python
claimants_data=pd.read_csv('claimants.csv')
claimants_data.head()
```

## 3. Initial Investigation

```python
claimants_data.shape
```

```python
claimants_data.isnull().sum()
```

## 4. Data Preprocessing/Preparation

```python
claimants_data.head(50)
```

```python
del claimants_data['CASENUM']
```

```python
claimants_data.head()
```

```python
claimants_data.dropna(axis=0,inplace=True)
```

```python
claimants_data.isnull().sum()
```

```python
claimants_data.shape
```

```python
X=claimants_data.iloc[:,1:]
y=claimants_data[['ATTORNEY']]
```

```python
X.head()
```

```python
y
```

## 5. Model Building

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=128,test_size=0.20)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train
```

## 6. Model Training

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dt_model=DecisionTreeClassifier(criterion='gini')
dt_model.fit(X_train,y_train)
```

## 7. Model Testing

```python
y_pred_train_data=dt_model.predict(X_train)#training test
```

```python
y_pred=dt_model.predict(X_test)#testing test
```

## 8. Model Evaluation

```python
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,roc_auc_score
```

### 8.1 Train Data

```python
confusion_matrix(y_train,y_pred_train_data)
```

```python
print(classification_report(y_train,y_pred_train_data))
```

```python
# Accuracy=100%
```

```python
model_3=DecisionTreeClassifier(max_depth=5).fit(X_train,y_train)
plt.figure(figsize=(16,10))
plot_tree(model_3,rounded=True,filled=True)
plt.show
```

```python
fpr,tpr,thresholds=roc_curve(y_train,y_pred_train_data)

auc=roc_auc_score(y_train,y_pred_train_data)
print(auc)

import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color='red',label='logit model(area=%0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1-True Negative Rate]')
plt.ylabel('True Positive Rate')
```

### 8.2 Test Data

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Accuracy=69%
```

```python
model_3=DecisionTreeClassifier(max_depth=5).fit(X_test,y_test)
plt.figure(figsize=(16,10))
plot_tree(model_3,rounded=True,filled=True)
plt.show
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_pred)
auc=roc_auc_score(y_test,y_pred)
print(auc)

plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
```
